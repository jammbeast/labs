#### БАЗА (1-5) задания

```python
import string;
from collections import Counter;
import math;

######################
file_path = "text.txt"
######################

with open(file_path, 'r', encoding='UTF-8') as file:
    text = file.read()
    cleanText = text.lower().translate(str.maketrans('','', string.punctuation)).replace(' ', '')
    charCounter = Counter(cleanText) 

    ########################################

    binGram_counts = {}
    for i in range(len(cleanText) -1):
        bingram = cleanText[i:i+2];
        if bingram in binGram_counts:
            binGram_counts[bingram] +=1;
        else:
            binGram_counts[bingram] = 1;
    ########################################
    totalChars = sum(charCounter.values()) 
    charProbs = {char : count / totalChars for char, count in charCounter.items()}

    entropy = -sum( p * math.log2(p) for p in charProbs.values())
    ########################################

    totalBinGrams = sum(binGram_counts.values())

    binGramProbs = {binGram : count / totalBinGrams for binGram, count in binGram_counts.items()}

    binGramEntropy = -sum( p * math.log2(p) for p in binGramProbs.values())
    #########################################
    # Для однобуквенных символов и бинграм при равноменном кодировании формула Длины кода : l = |log2(m).

    lengChars = math.ceil(math.log2(len(charProbs)))

    lengBin = math.ceil(math.log2(len(binGramProbs)))
    ########################################

    
    #D0 = 1 - log2(l)/l

    #Dp = 1 - H / log2(m)

    D0_chars = 1 - math.log2(len(charProbs)) / lengChars

    Dp_chars = 1 - entropy / math.log2(len(charProbs)) 


    D0_bingrams = 1 - math.log2(len(binGramProbs)) / lengBin

    Dp_bingrams = 1 - binGramEntropy / math.log2(len(binGramProbs))

    #######################
    



print(cleanText)
print(Counter(cleanText))
print(f"{binGram_counts}\n")

print(f"Вертояности символов:{charProbs}\n")
print(f"Энтропия для символов: {entropy} бит\n")
print(f"Энтропия для бинграм: {binGramEntropy} бит\n")

print(f"Длина кода для символов:{lengChars} бит, для бингамм :{lengBin} бит")

print(f"Симолы : избыточность длины кода: {D0_chars}, избыточность сообщения : {Dp_chars} \n")
print(f"Бинграмы: избыточность длины кода{D0_bingrams}, избыточность сообщения :{Dp_bingrams}\n")
```

```python
import string
from collections import Counter
import math

######################
file_path = "text.txt"
######################

with open(file_path, 'r', encoding='UTF-8') as file:
    text = file.read()
    cleanText = text.lower().translate(str.maketrans('', '', string.punctuation)).replace(' ', '')
    charCounter = Counter(cleanText)

    least_frequent = [char for char, _ in charCounter.most_common()[:-math.ceil(len(charCounter) * 0.2) - 1:-1]]
    text_without_min = ''.join([char for char in cleanText if char not in least_frequent])
    charCounterLeast = Counter(text_without_min)
#############################################################################################
    binGram_counts = {}
    for i in range(len(text_without_min) - 1):
        bingram = text_without_min[i:i+2]
        if bingram in binGram_counts:
            binGram_counts[bingram] += 1
        else:
            binGram_counts[bingram] = 1
    ########################################
    totalChars = sum(charCounterLeast.values())

    charProbs = {char: count / totalChars for char, count in charCounterLeast.items()}

    entropy = -sum(p * math.log2(p) for p in charProbs.values())
    ########################################

    totalBinGrams = sum(binGram_counts.values())

    binGramProbs = {binGram: count / totalBinGrams for binGram, count in binGram_counts.items()}

    binGramEntropy = -sum(p * math.log2(p) for p in binGramProbs.values())
    #########################################
    # Для однобуквенных символов и бинграм при равноменном кодировании формула Длины кода : l = |log2(m).

    lengChars = math.ceil(math.log2(len(charProbs)))

    lengBin = math.ceil(math.log2(len(binGramProbs)))
    ########################################
    # D0 = 1 - H/l

    # Dp = 1 - H / log2(m)

    D0_chars = 1 - entropy / lengChars

    Dp_chars = 1 - entropy / math.log2(len(charProbs))

    D0_bingrams = 1 - binGramEntropy / lengBin

    Dp_bingrams = 1 - binGramEntropy / math.log2(len(binGramProbs))

    #######################

print(text_without_min)
print(Counter(text_without_min))
print(f"{binGram_counts}\n")
print(f"Вероятности символов: {charProbs}\n")
print(f"Энтропия для символов: {entropy} бит\n")
print(f"Энтропия для бинграм: {binGramEntropy} бит\n")

print(f"Длина кода для символов: {lengChars} бит, для бинграм: {lengBin} бит")

print(f"Символы: избыточность длины кода: {D0_chars}, избыточность сообщения: {Dp_chars} \n")
print(f"Бинграмы: избыточность длины кода: {D0_bingrams}, избыточность сообщения: {Dp_bingrams}\n")

######

```
